```python

anzahl = int(input("Wie viele Sch端ler gibt es? "))  
namen = []
noten = []

for i in range(anzahl):
    name = input(f"Name von Sch端ler {i+1}: ")
    note = int(input(f"Note von {name}: "))
    namen.append(name)
    noten.append(note)

print("\nAlle Sch端ler:")
for i in range(anzahl):
    print(f"{namen[i]} -> {noten[i]}") 

#Durchschnittsnote
durchschnitt = sum(noten) / len(noten)
print("Durchschnittsnote:", durchschnitt)

#Beste Note
beste_note = min(noten)
bester_schueler = namen[noten.index(beste_note)]
print("Beste Note:", bester_schueler, "mit", beste_note)

#Schlechteste Note
schlechteste_note = max(noten)
schlechtester_schueler = namen[noten.index(schlechteste_note)]
print("Schlechteste Note:", schlechtester_schueler, "mit", schlechteste_note)
```
