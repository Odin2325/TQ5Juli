```python
import os
import datetime
import psutil

def is_locked(path):
    """Prüfen, ob ein Prozess die Datei geöffnet hat (schonender)."""
    for proc in psutil.process_iter(['pid', 'name']):
        try:
            for f in proc.open_files():
                if f.path == path:
                    return True
        except (psutil.AccessDenied, psutil.NoSuchProcess):
            continue
    return False


# Programstart
print("Cleanup gestartet!")
print("Gestartet von: Hannes am", datetime.datetime.today().strftime("%d-%m-%Y"))

# Pfad zum Temp-Ordner
ordner = r"C:\Users\nicoa\Desktop\TQ5Juli\12Sep\SysAdmin"
dateien = os.listdir(ordner)

# Filter
gefilterte_dateien = [f for f in dateien if f.endswith('.tmp') or f.endswith('.log')]

# Schleife mit Ausgabe
print(f"Dateien im Ordner {ordner}:")
for datei in gefilterte_dateien:
    print(datei)

# Löschvorgang
for datei in gefilterte_dateien:
    dateipfad = os.path.join(ordner, datei)
    if is_locked(dateipfad):
        print(f"Übersprungen (in Benutzung): {datei}")
        continue

    try:
        os.remove(dateipfad)
        print(f"Gelöscht: {datei}")
    except Exception as e:
        print(f"Konnte nicht gelöscht werden: {datei}. Grund: {e}")

# Programmende
print("Cleanup beendet!")
```

```python
import os
import datetime
 
def cleanup_programm():
    print("Cleanup-Programm gestartet!")
    print(f"Autor: Dein Name")
    print(f"Datum: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print("="*50)
 
    # Benutzer nach Ordner fragen
    target_dir = input("Bitte den Pfad zum Ordner eingeben: ").strip()
    if not os.path.exists(target_dir):
        print("Ordner existiert nicht!")
        return
 
    geloescht = 0
    logfile = os.path.join(target_dir, "cleanup.log")
 
    with open(logfile, "a", encoding="utf-8") as log:
        log.write(f"\n--- Cleanup gestartet am {datetime.datetime.now()} ---\n")
 
        for root, dirs, files in os.walk(target_dir):
            for file in files:
                if file.endswith((".tmp", ".log")):
                    filepath = os.path.join(root, file)
                    try:
                        os.remove(filepath)
                        print(f"{file} gelöscht.")
                        log.write(f"Gelöscht: {filepath}\n")
                        geloescht += 1
                    except Exception as e:
                        print(f"Fehler bei {file}: {e}")
                        log.write(f"Fehler: {filepath} → {e}\n")
 
        log.write(f"--- {geloescht} Dateien gelöscht ---\n")
 
    print("="*50)
    print(f"Cleanup beendet – {geloescht} Dateien gelöscht.")
    print(f"Log gespeichert in: {logfile}")
 
def log_anzeigen(target_dir):
    logfile = os.path.join(target_dir, "cleanup.log")
    if os.path.exists(logfile):
        print("\n--- Inhalt von cleanup.log ---")
        with open(logfile, "r", encoding="utf-8") as log:
            print(log.read())
    else:
        print("Noch kein Logfile vorhanden.")
 
# Hauptmenü
if __name__ == "__main__":
    while True:
        print("\n=== Menü ===")
        print("1. Cleanup starten")
        print("2. Log anzeigen")
        print("3. Beenden")
        wahl = input("Bitte wählen (1-3): ")
 
        if wahl == "1":
            cleanup_programm()
        elif wahl == "2":
            ordner = input("Ordner für das Log eingeben: ").strip()
            log_anzeigen(ordner)
        elif wahl == "3":
            print("Programm beendet.")
            break
        else:
            print("Ungültige Eingabe.")
```
