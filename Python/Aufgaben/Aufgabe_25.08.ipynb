```python
### Einkaufsliste mit Preisberechnung
def einkaufsliste():
    """
    Startet eine einfache Einkaufsliste-Abfrage.

    Der Benutzer kann Produkte eingeben, die im Sortiment liegen.
    Wenn ein Produkt im Dictionary vorhanden ist, wird sein Preis
    zu den Gesamtkosten addiert. Mit "Ende" wird die Eingabe beendet.

    Returns:
        float: Gesamtkosten aller eingekauften Produkte.
    """
    produkte = {
        "Apfel": 0.5,
        "Banane": 0.8,
        "Milch": 1.2,
        "Brot": 2.0,
        "Käse": 2.5
    }

    gesamtkosten = 0.0

    while True:
        produkt = input("Produkt eingeben (oder 'Ende' zum Beenden): ")
        
        if produkt.lower() == "ende":
            break
        
        if produkt in produkte:
            gesamtkosten += produkte[produkt]
        else:
            print(f"{produkt} ist nicht im Sortiment.")

    print(f"\nGesamtkosten: {gesamtkosten:.2f} €")
    return gesamtkosten
```

```python
##Blackjack Spiel gegen den Computer (Dealer)
import random

def blackjack():
    """
    Startet ein einfaches Blackjack-Spiel gegen den Computer (Dealer).

    Ablauf:
        - Spieler erhält zwei zufällige Karten.
        - Dealer erhält zwei zufällige Karten (eine wird angezeigt).
        - Spieler entscheidet: "Hit" (neue Karte) oder "Stand" (keine Karte mehr).
        - Dealer zieht Karten, bis er mindestens so viele Punkte wie der Spieler hat.
        - Am Ende wird entschieden, ob Spieler oder Dealer gewonnen hat.

    Hinweise:
        - Kartenwerte werden über die Hilfsfunktion `punkte_berechnen()` berechnet.
        - Ass zählt standardmäßig 11 Punkte.

    Returns:
        None
    """
    spielstappel = [2,3,4,5,6,7,8,9,10,'bube','dame','koenig','ass']*4
    karten_werte = {
        'bube' : 10,
        'dame' : 10,
        'koenig' : 10,
        'ass' : 11
    }

    spieler_karten = [random.choice(spielstappel), random.choice(spielstappel)]
    dealer_karten = [random.choice(spielstappel), random.choice(spielstappel)]

    print('Meine Karten sind: ', spieler_karten)
    print('Dealer Karte:', dealer_karten[0])

    # Spieler Schleife
    while True:
        eingabe = input('Möchten Sie ein "Hit" oder "Stand"? ')
        if eingabe.lower() == "hit":
            print('Sie haben Hit ausgewählt.')
            spieler_karten.append(random.choice(spielstappel))
            print(spieler_karten)
        elif eingabe.lower() == "stand":
            print('Sie haben Stand ausgewählt.')
            break
        else:
            print('Das haben wir nicht verstanden. Bitte nur Hit oder Stand eingeben!')

    # Punkte berechnen mit Methode aus Aufgabe 3
    spieler_punkte = punkte_berechnen(spieler_karten, karten_werte)

    if spieler_punkte > 21:
        print('Sie haben verloren.')
    else:
        dealer_punkte = punkte_berechnen(dealer_karten, karten_werte)
        print("Dealer Karten:", dealer_karten)

        while dealer_punkte < spieler_punkte:
            neue_karte = random.choice(spielstappel)
            dealer_karten.append(neue_karte)
            print("Dealer Karten:", dealer_karten)
            dealer_punkte = punkte_berechnen(dealer_karten, karten_werte)

        if dealer_punkte > 21 or spieler_punkte > dealer_punkte:
            print('Der Dealer hat verloren! Wir haben gewonnen!')
        elif dealer_punkte == spieler_punkte:
            print('Unentschieden!')
        else:
            print('Der Dealer hat gewonnen...')
```

```python
#Aufgabe 3 – Punkte berechnen
def punkte_berechnen(karten, karten_werte):
    """
    Berechnet die Gesamtpunktzahl einer Kartenliste im Blackjack.

    Jede Karte wird anhand des Dictionary `karten_werte` ausgewertet.
    Zahlenkarten (int) werden direkt addiert, Bildkarten/Ass nach Dictionary.

    Args:
        karten (list): Liste von Karten (z. B. ['bube', 3, 8]).
        karten_werte (dict): Dictionary mit Werten für Bildkarten/Ass.

    Returns:
        int: Gesamtpunktzahl der Karten.
    
    Beispiel:
        >>> punkte_berechnen(['bube', 3, 8], {'bube': 10, 'dame': 10, 'koenig': 10, 'ass': 11})
        21
    """
    punkte = 0
    for karte in karten:
        if isinstance(karte, str):   # Figur oder Ass
            punkte += karten_werte[karte]
        else:                        # normale Zahl
            punkte += karte
    return punkte
```
