Aufgabenstellung: Konsolenprogramm zur Personaldatenverwaltung mit Datenbankanbindung 
Ziel: Entwickeln Sie ein Konsolenprogramm zur Erfassung und Verwaltung von Personaldaten. Die Daten sollen über die Konsole eingegeben und in einer Datenbank gespeichert sowie wieder ausgelesen werden können. 

Funktionale Anforderungen: 

1. Datenerfassung über die Konsole: 
o Das Programm soll den Benutzer auffordern, folgende Daten einzugeben: 
▪ Vorname 
▪ Nachname 
▪ Geburtsdatum (Format: TT.MM.JJJJ) 
▪ Anschrift 
▪ Körpergröße (in cm) 
▪ Augenfarbe 

2. Datenbankanbindung: 
o Die Daten sollen in einer relationalen Datenbank gespeichert werden (z. B. SQLite). 
o Die Datenbank soll eine Tabelle personen enthalten mit folgenden Spalten: 
▪ id (Primärschlüssel, automatisch) 
▪ vorname 
▪ nachname 
▪ geburtsdatum 
▪ anschrift 
▪ koerpergroesse 
▪ augenfarbe 

3. Funktionen des Programms: 
o Datensatz hinzufügen: Benutzer gibt die Daten ein, und sie werden gespeichert. 
o Alle Datensätze anzeigen: Alle gespeicherten Personen werden tabellarisch ausgegeben. 
o Programmmenü: Das Programm zeigt ein Menü mit Optionen: 
▪ 1 – Neue Person hinzufügen 
▪ 2 – Alle Personen anzeigen 
▪ 3 – Beenden 

4. Validierung: 
o Körpergröße muss eine positive Zahl sein. 
o Geburtsdatum darf nicht in der Zukunft liegen. 
o Alle Felder müssen ausgefüllt sein.

```python
import sqlite3

# create database und connect

db = sqlite3.connect("app.db")

# create table

db.execute("""
create table if not exists personen (
id int primary key autocrement,
vorname text,
nachname text,
geburstagdatum text,
anschrift text,
koerpergroesse int,
augenfarbe text               
)
""")

db.commit()

# funktionen

def Datensatz_hinzufügen():
    vorname = input("Vorname : ")
    nachname = input("Nachname : ")
    geburstagdatum = input("Geburstagdatum : ")
    anschrift = input("Anschrift : ")
    koerpergroesse = int(input("Körpergröße : "))
    augenfarbe = input("Augenfarbe : ")


db.execute("""
insert into personen()
values(Natalie, Müller, 01.10.2000, wegstr 18b 12345 Berlin, 175, brawn)
""")
db.commit()

print("die Daten sind schon gespeichert")


def Datensatz_anzeigen():
    db.execute("select * from personen")
    people = db.fetchall()
    
    for items in people:
        print(items)


def Menue():
    while True:
        print ("*************Personaldatenverwaltung**************")
        print("1 Datensatz hinzufügen")
        print("2 Datensatz anzeigen")
        print("3 Programm Beenden")
        print("***************************************************")

        w = input("Geben Sie Ihre Wahl ein : ")

        if w == "1":
            Datensatz_hinzufügen()
        elif w == "2":
            Datensatz_anzeigen()
        elif w == "3":
            break
        else :
            print("Error! Geben Sie Nummer 1 oder 2 oder 3 ")



db.close()




            




```
