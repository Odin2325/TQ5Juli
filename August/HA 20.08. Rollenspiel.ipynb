Aufgabe 2 Rollenspielchen:
In einigen Rollenspielen kämpfen die Figuren der Spieler*innen gegen nicht von
Spieler*innen gesteuerte Figuren, die in der Regel die Monster oder Schurken sind.
Manchmal besteht ein Kampf darin, dass jede Figur mithilfe von Würfeln einen Zufallswert
generiert, der dann vom Gesundheitswert des Gegners abgezogen wird. Sobald die
Gesundheit einer Figur null erreicht, stirbt die Figur, oder sie verliert das Spiel.
In dieser Aufgabe bringen wir diese Interaktion auf den Punkt. Ein Held und ein Monster
beginnen mit dem gleichen Integritätspunktestand. Wenn die Heldin oder der Held am Zug
ist, generiert sie bzw. er einen Zufallswert, der vom Gesundheitswert des Monsters
abgezogen wird. Wenn der Gesundheitswert des Monsters größer als null ist, kommt es an
die Reihe und greift den Helden bzw. die Heldin an. Solange sowohl der Held bzw. die Heldin
als auch das Monster über Gesundheitswerte größer als null verfügen, wird der Kampf
fortgesetzt.
Programmieraufgabe: Schreiben von Code zum Implementieren der Spielregeln
Hier sind die Regeln für das Kampfspiel, das Sie in Ihrem Codeprojekt implementieren sollen:
1. Sie müssen die while-Anweisung als äußere Spielschleife verwenden.
2. Der Held bzw. die Heldin und das Monster beginnen mit 10 Gesundheitspunkten.
a. Eine variable held und monster erstellen, beide gleich 10.
3. Alle Angriffe haben einen Wert zwischen 1 und 10. Diese sollen zuffaellig generiert
werden.
a. randomint(1,10)
4. Der Held bzw. die Heldin greift zuerst an.
5. Geben Sie die Integritätsmenge, die das Monster verloren hat, und seine
verbleibende Integrität aus.
6. Wenn die Integrität des Monsters größer als 0 ist, kann es den Helden angreifen.
7. Geben Sie die Integritätsmenge, die der Held verloren hat, und seine verbleibende
Integrität aus.
8. Führen Sie diese Abfolge von Angriffen fort, bis entweder die Integrität des Monsters
oder die des Helden null oder weniger ist.
9. Geben Sie den Sieger aus.

```python
import random

# Grundwerte der Figuren
hp_pmc = 100
hp_npc1 = 50
hp_npc2 = 50

str_pmc = 50
str_npc1 = 25
str_npc2 = 25

def_pmc = 20  
def_npc1 = 30
def_npc2 = 30

# Spezialattacken-Zähler
pmcsp = 0

# Haupt-Spielschleife
while hp_pmc > 0 and (hp_npc1 > 0 or hp_npc2 > 0):
    
    # 1. Statusanzeige
    print("-" * 30)
    print(f"Deine HP: {hp_pmc}")
    print(f"NPC1 HP: {hp_npc1}")
    print(f"NPC2 HP: {hp_npc2}")
    print("-" * 30)

    # Erhöhen des Spezialattacken-Zählers
    pmcsp += 1

    # 2. Spieler-Aktion
    # Überprüfen, ob Spezialattacke bereit ist
    if pmcsp >= 10:
        print("Deine Spezialattacke ist bereit!")
        aktion = input("Wähle deine Aktion: (1) Angriff, (2) Spezialattacke: ")
        
        if aktion == '2':
            spezial_aktion = input("Wähle Spezialattacke: (1) Schaden, (2) Heilung: ")
            
            if spezial_aktion == '1':
                # Spezial-Schadensangriff
                if hp_npc1 > 0:
                    hp_npc1 -= 20
                    print("Spezialattacke: NPC1 erleidet 20 HP Schaden!")
                if hp_npc2 > 0:
                    hp_npc2 -= 20
                    print("Spezialattacke: NPC2 erleidet 20 HP Schaden!")
            
            elif spezial_aktion == '2':
                # Spezial-Heilungsangriff
                hp_pmc += 25
                print("Spezialattacke: Du hast dich um 25 HP geheilt!")
            
            else:
                print("Ungültige Spezialattacken-Auswahl.")
            
            # Spezialattacken-Zähler zurücksetzen
            pmcsp = 0
            
        elif aktion == '1':
            # Normaler Angriff 
            ziel = input("Welchen NPC möchtest du angreifen? (1 oder 2): ")
            # Angriffslogik
            if ziel == '1' and hp_npc1 > 0:
                schaden = random.randint(0, str_pmc)
                if schaden > def_npc1:
                    differenz = schaden - def_npc1
                    hp_npc1 -= differenz
                    print(f"Du hast NPC1 {differenz} Schaden zugefügt!")
                else:
                    print("Dein Angriff war nicht stark genug!")
            elif ziel == '2' and hp_npc2 > 0:
                schaden = random.randint(0, str_pmc)
                if schaden > def_npc2:
                    differenz = schaden - def_npc2
                    hp_npc2 -= differenz
                    print(f"Du hast NPC2 {differenz} Schaden zugefügt!")
                else:
                    print("Dein Angriff war nicht stark genug!")
            else:
                print("Ungültiges Ziel oder Ziel ist bereits besiegt.")
        
        else:
            print("Ungültige Aktionsauswahl.")
            
    else: # Wenn Spezialattacke nicht bereit ist
        ziel = input("Welchen NPC möchtest du angreifen? (1 oder 2): ")
        
        if ziel == '1' and hp_npc1 > 0:
            schaden = random.randint(0, str_pmc)
            if schaden > def_npc1:
                differenz = schaden - def_npc1
                hp_npc1 -= differenz
                print(f"Du hast NPC1 {differenz} Schaden zugefügt!")
            else:
                print("Dein Angriff war nicht stark genug!")
        elif ziel == '2' and hp_npc2 > 0:
            schaden = random.randint(0, str_pmc)
            if schaden > def_npc2:
                differenz = schaden - def_npc2
                hp_npc2 -= differenz
                print(f"Du hast NPC2 {differenz} Schaden zugefügt!")
            else:
                print("Dein Angriff war nicht stark genug!")
        else:
            print("Ungültiges Ziel oder Ziel ist bereits besiegt.")
        
    # 3. NPC-Angriffe
    # NPC1 greift an
    if hp_npc1 > 0:
        schaden_npc1 = random.randint(0, str_npc1)
        if schaden_npc1 > def_pmc:
            differenz = schaden_npc1 - def_pmc
            hp_pmc -= differenz
            print(f"NPC1 greift dich an und fügt dir {differenz} Schaden zu!")
        else:
            print("NPC1's Angriff war nicht stark genug!")
            
    # NPC2 greift an
    if hp_npc2 > 0:
        schaden_npc2 = random.randint(0, str_npc2)
        if schaden_npc2 > def_pmc:
            differenz = schaden_npc2 - def_pmc
            hp_pmc -= differenz
            print(f"NPC2 greift dich an und fügt dir {differenz} Schaden zu!")
        else:
            print("NPC2's Angriff war nicht stark genug!")

# Ende des Spiels
print("-" * 30)
if hp_pmc <= 0:
    print("Du wurdest besiegt! Spiel vorbei.")
else:
    print("Glückwunsch! Du hast die NPCs besiegt!")
```
