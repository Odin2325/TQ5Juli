```python
def einkaufsliste_erstellen():
    """
    Diese Funktion erstellt eine Einkaufsliste, die vom Benutzer befuellt werden muss.

    Returns:
        list: Eine Liste mit produkte.

    Raises:
        KeyboardInterruptError: Benutzer hat esc taste gedrueckt statt quit einzutippen.
    """

    einkaufliste = []
    eingabe = ''

    while eingabe != 'quit':
        eingabe = input('Was moechten Sie einkaufen? (quit zum beenden) ')
        if eingabe != 'quit':
            einkaufliste.append(eingabe)

    print('Ihre Einkaufsliste: ', einkaufliste)
    return einkaufliste
```

```python
#Aufgabe 2 BlackJack

import random

def blackjack_spiel():
    # Erstellen des Decks:
    # 4x die Werte 2-9
    # 16x der Wert 10 (4x Zehn, 4x Bube, 4x Dame, 4x König)
    # 4x der Wert 11 (Ass)
    deck = [2, 3, 4, 5, 6, 7, 8, 9] * 4 + [10] * 16 + [11] * 4

    # Mischen des Decks
    random.shuffle(deck)

    # Austeilen der Karten
    spielerkarte1 = deck.pop()
    spielerkarte2 = deck.pop()

    dealerkarte1 = deck.pop()
    dealerkarte2 = deck.pop()

    spielerpunkte = (spielerkarte1 + spielerkarte2)
    dealerpunkte = (dealerkarte1 + dealerkarte2)

    print(f"Spieler hat die Karten: {spielerkarte1} und {spielerkarte2}")
    print(f"Dealer hat die Karten: {dealerkarte1}")

    while spielerpunkte < 21:
        runde1 = input("1 für hit und 2 für Stand: ")

        if runde1 == "1":
            spielerHit1 = deck.pop()
            spielerpunkte = (spielerpunkte + spielerHit1)
            print(f"Spieler Zieht eine weitere Karte: {spielerHit1}\nSpieler Gesamtpunkte: {spielerpunkte}")
        elif runde1 == "2":
            break
        else:
            print(f"Ungültige Eingabe, bitte 1 oder 2 eingeben.")

    if spielerpunkte > 21:
        print(f"Spieler hat {spielerpunkte} Punkte! Dealer Gewinnt")
        return "Dealer"
    elif spielerpunkte == 21:
        print("Blackjack! Spieler gewinnt.")
        return "Spieler"
    else:
        print(f"Dealer dreht Karte 2 um, die eine {dealerkarte2} ist und somit: {dealerpunkte} Punkte hat")

        while dealerpunkte < 17:
            dealerHit1 = deck.pop()
            dealerpunkte = (dealerpunkte + dealerHit1)
            print(f"Dealer Zieht eine weitere Karte.\nDealer hat eine {dealerHit1} gezogen und somit {dealerpunkte} Gesamtpunkte")

        if dealerpunkte > 21:
            print(f"Dealer hat {dealerpunkte} Punkte. Spieler Gewinnt!")
            return "Spieler"
        elif dealerpunkte < spielerpunkte:
            print(f"Dealer hat {dealerpunkte} Punkte und Spieler hat {spielerpunkte} Punkte")
            print("+" * 35)
            print("!!!!!!!!!!SPIELER GEWINNT!!!!!!!!!!")
            print("+" * 35)
            return "Spieler"
        elif spielerpunkte < dealerpunkte:
            print(f"Dealer hat {dealerpunkte} Punkte und Spieler hat {spielerpunkte} Punkte")
            print("+" * 35)
            print("!!!!!!!!!!DEALER GEWINNT!!!!!!!!!!")
            print("+" * 35)
            return "Dealer"
        else:
            print("\n!!!!UNENTSCHIEDEN KEINER GEWINNT!!!!!")
            return "Unentschieden keiner"

# Hauptprogramm
gewinner = blackjack_spiel()
print(f"{gewinner} hat gewonnen!")
```
