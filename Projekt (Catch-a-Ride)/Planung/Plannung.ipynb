1. Funktion: Fahrzeug mieten

function rent_vehicle(customer_id, vehicle_id, start_date, end_date):
    prüfe ob Fahrzeug verfügbar
    wenn nicht verfügbar:
        gib Fehlermeldung zurück
    prüfe ob Kunde aktiven Mietvertrag hat
    wenn ja:
        gib Fehlermeldung zurück
    berechne Mietzeitraum und Kosten
    erstelle Mietvertrag in Rental-Tabelle
    setze Fahrzeugstatus auf "Rented"
    gib Bestätigung zurück

2. Funktion: Fahrzeug zurückgeben
function return_vehicle(rental_id, km_ende, fuel_level):
    hole Mietvertrag anhand rental_id
    berechne gefahrene Kilometer
    berechne Zusatzkosten (Tanken, Verspätung, Schäden)
    erstelle Rechnung
    setze Fahrzeugstatus auf "Available"
    markiere Mietvertrag als abgeschlossen
    gib Rechnung an Kunden zurück

3. Funktion: Zahlung erfassen
function record_payment(rental_id, amount, payment_method):
    prüfe ob Mietvertrag existiert
    wenn nicht → Fehlermeldung
    prüfe ob Betrag korrekt
    wenn falsch → Fehlermeldung
    speichere Zahlung in Payment-Tabelle
    setze Mietvertrag auf "bezahlt"
    gib Zahlungsbestätigung zurück

4. Funktion: Fahrzeug registrieren
function register_vehicle(vehicle_data):
    prüfe ob Fahrzeug-ID bereits existiert
    wenn ja → Fehlermeldung
    speichere Fahrzeugdaten in Vehicle-Tabelle
    setze Status auf "Available"
    gib Bestätigung zurück

2. Fahrzeug zurückgeben

Start → Fahrzeug abgeben → Kilometerstand eingeben → Tankfüllung prüfen → Kosten berechnen → Rechnung erstellen → Zahlung erfassen → Status „Available“ setzen → Ende

3. Kunde registrieren

Start → Kundendaten eingeben → prüfen ob Kunde existiert

[ja] → Fehlermeldung

[nein] → Daten speichern → Ende

4. Zahlung erfassen

Start → Zahlungsdaten eingeben → prüfen ob Betrag korrekt

[falsch] → Fehlermeldung

[korrekt] → speichern → Mietvertrag auf „bezahlt“ setzen → Ende

5. Fahrzeug warten

Start → Fahrzeugstatus „In Wartung“ setzen → Reparatur durchführen → Status auf „Available“ setzen → Ende
