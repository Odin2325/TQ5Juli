```python

personen_db = {}            # Hier speichern wir alle Personen
naechste_id = 1             # Startwert für die erste Person-ID

def neue_person():
    global naechste_id      # Damit wir die ID-Nummer verändern können, global damit ausserhalb der funktion arbeitet
    vorname = input("Vorname: ")              
    nachname = input("Nachname: ")            
    geburtsdatum = input("Geburtsdatum (TT.MM.JJJJ): ") 
    anschrift = input("Anschrift: ")        
    koerpergroesse = input("Körpergröße (cm): ") 
    augenfarbe = input("Augenfarbe: ")       

    # Daten in einem Dictionary speichern
    person = {
        "vorname": vorname,       
        "nachname": nachname,         
        "geburtsdatum": geburtsdatum, 
        "anschrift": anschrift,       
        "koerpergroesse": koerpergroesse, 
        "augenfarbe": augenfarbe      
    }

    # Die Person mit der aktuellen ID in der "Datenbank" speichern
    personen_db[naechste_id] = person
    print(f"Person mit ID {naechste_id} gespeichert.\n")

    naechste_id += 1        #auto increment primary key 


# Funktion zum Anzeigen aller gespeicherten Personen
def alle_personen_anzeigen():
    if not personen_db:                    # Wenn die Datenbank leer ist
        print("Keine Personen gespeichert.\n")
        return                             
    for id, person in personen_db.items(): 
        print(f"ID: {id}")                 
        for key, value in person.items():  
            print(f"  {key}: {value}")     
        print()                            # Leerzeile 


# Hauptmenü für das Programm
def menue():
    while True:                           # Endlosschleife, bis man "Beenden" wählt
        print("1 – Neue Person hinzufügen")
        print("2 – Alle Personen anzeigen")
        print("3 – Beenden")

        wahl = input("Wähle eine Option(1,2,3): ")  # Benutzer gibt Menüauswahl ein

        if wahl == "1":                   
            neue_person()                 
        elif wahl == "2":                 
            alle_personen_anzeigen()      
        elif wahl == "3":                 
            print("Programm beendet.")   
            break                         
        else:                            
            print("Ungültige Eingabe.\n") # Fehlermeldung


# Hier startet das Programm mit dem Menü
menue()
```
