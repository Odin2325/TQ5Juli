```python
from Bank import Bank
from KundeDatei import Kunde
from BankkontoDatei import Bankkonto

meine_bank = Bank("Parkbank", "Stadtpark", "NRW", "Deutschland" )
#meine_konto_liste = Bankkonto("Max","Mustermann","2","10000000", ) 
```

```python
# Startmenü

def start_menue ():
    while True:
        print("Willkommen im Startmenü!\n")
        print("1. Kunde werden\n")
        print("2. Login\n")
        print("3. Beenden\n")
        auswahl = input("Auswahl: ")

        if auswahl == "1":
            kunde_werden()
        elif auswahl == "2":
            login()
        elif auswahl == "3":
            print("Programm wird beendet.")
            break
        else:
            print("Ungültige Auswahl. Bitte erneut versuchen.")

# Kunde werden
def kunde_werden():
    vorname = input("Vorname: ")
    nachname = input("Nachname: ")
    adresse = input("Adresse: ")
    telefonnummer = input("Telefonnummer: ")
    email = input("E-Mail: ")
    alter = int(input("Alter: ")) 

    #meine_bank.kunde_erstellen(vorname, nachname, adresse, telefonnummer, email, alter) 
    #neuer_kunde = Kunde(vorname, nachname, adresse, telefonnummer, email, alter)
    neuer_kunde = meine_bank.kunde_erstellen(vorname, nachname, adresse, telefonnummer, email, alter)

    print(f"Kunde {vorname} {nachname} wurde erfolgreich erstellt.")
    kunden_menue_neukunde(neuer_kunde)
    for kunde in meine_bank.kunden:
        if kunde.vorname == vorname and kunde.nachname == nachname:
            kunden_menue_neukunde(kunde)
            break

# Kunden Login
def login():
    vorname = input("Vorname: ")
    nachname = input("Nachname: ")
    for kunde in meine_bank.kunden:
        if kunde.vorname == vorname and kunde.nachname == nachname:
            print(f"Willkommen zurück, {vorname} {nachname}!")
            kunden_menue_neukunde(kunde)
            return
        else:
            print("Kunde nicht gefunden. Bitte erneut versuchen.")

# Kundenmenü Neukunde
def kunden_menue_neukunde(kunde: Kunde):
    while True:
        print(f"\nKundenmenü für {kunde.vorname} {kunde.nachname}")
        print("1. Details ausgeben")
        print("2. Bankonten erstellen")
        print("3. Abmelden")
        auswahl = input("Auswahl: ")
        if auswahl == "1":
            kunde.detail_kunde()
        elif auswahl == "2":
            meine_bank.bankkonto_erstellen(kunde)
        elif auswahl == "3":
            print("Abmeldung erfolgreich.")
            break
        else:
            print("Ungültige Auswahl. Bitte erneut versuchen.")
    
# Kundenmenü Bestandskunde
def konto_menue(konto, kunde):
    while True:
        print(f"\n==== Konto Menü (Konto {konto.kontonummer}) für {kunde.vorname} {kunde.nachname} ====")
        print("1. Einzahlen")
        print("2. Auszahlen")
        print("3. Überweisen")
        print("4. Kontoauszug")
        print("5. Kontodetails")
        print("6. Zurück zum Kundenmenü")
        wahl = input("Auswahl: ")

        if wahl == "1":
            betrag = float(input("Betrag zum Einzahlen: "))
            konto.einzahlen(betrag)
        elif wahl == "2":
            betrag = float(input("Betrag zum Auszahlen: "))
            konto.auszahlen(betrag)
        elif wahl == "3":
            empfänger_nummer = int(input("Empfängerkontonummer: "))
            empfänger_konto = None
            for k in meine_bank.konten:
                if k.kontonummer == empfänger_nummer:
                    empfänger_konto = k
                    break
            if empfänger_konto:
                betrag = float(input("Betrag zum Überweisen: "))
                konto.überweisen(empfänger_konto, betrag)
            else:
                print("Empfängerkonto nicht gefunden.")
        elif wahl == "4":
            konto.kontoauszug()
        elif wahl == "5":
            konto.kontodetails()
        elif wahl == "6":
            break
        else:
            print("Ungültige Auswahl. Bitte erneut versuchen.")
```
